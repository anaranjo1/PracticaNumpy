#### Carga de los datos

```python
import numpy as np

#cargamos los datos
gasto = np.genfromtxt('data/gov_10a_exp_page_linear.csv', delimiter=',', skip_header=1, dtype = str, encoding='utf-8')

print(gasto.shape) #observamos su estructura o forma.
print(type(gasto)) #comprobamos que es un array de numpy.
```

#### PRIMER Y SEGUNDO INDICADOR
En primer lugar vamos a filtrar los datos pertenecientes al año 2020 y vamos a buscar el valor máximo y mínimo de este array.

```python
#filtramos los datos para el año 2020

filtro2020 = np.asarray(['2020'])
#la función "np.in1d" permite comprobar si un elemento de un array está presente en otro.
gasto2020 = gasto[np.in1d(gasto[:, 8], filtro2020)] 

#creamos un array con la columna que contiene los datos del gasto público del año 2020
gasto2020_gdp = gasto2020[:,9].astype(float)
print(gasto2020_gdp)

#obtenemos el valor máximo de gasto público sobre el PIB
print("El valor máximo es de: " + str(np.max(gasto2020_gdp)))

#obtenemos el valor mínimo de gasto público sobre el PIB
print("El valor mínimo es de: " + str(np.min(gasto2020_gdp)))
```

#### TERCER INDICADOR

A continuación vamos a obtener un array con la evolución del gasto público de España como porcentaje del PIB y generaremos un gráfico.

```python
import numpy as np
from matplotlib import pyplot as plt #usamos matplotlib para poder generar el gráfico

filtro_pais = np.asarray(['ES'])
gasto_ES = gasto[np.in1d(gasto[:, 7], filtro_pais)]
print(gasto_ES)

#creamos un array con las columnas que contienen los datos pertenecientes al gasto como porcentaje del PIB y el año de esta observación
gasto_ES_evolucion = gasto_ES[:,8:10].astype(float)
print(gasto_ES_evolucion)

#procedemos a crear el gráfico
x, y = gasto_ES_evolucion.T #transponemos
plt.plot(x,y)
#añadimos un título a la gráfica
plt.title(label = "Evolución del gasto público como porcentaje del PIB en España", fontsize = 10)
#añadimos etiquetas
plt.xlabel('Años')
plt.ylabel('Gasto público como porcentaje del PIB (%)')
plt.show()
```

En el gráfico se puede observar la tendencia descendente del gasto público en España desde 2012 hasta 2017. Adicionalmente, resulta notorio destacar el aumento del gasto público durante el año 2020, causado principalmente por la coyuntura generada por la pandemia de la COVID-19 y la necesidad de aplicar políticas fiscales expansivas.
